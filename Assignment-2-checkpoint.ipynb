```python
#2. Filtering.
cylinder_array = df['cylinders'].to_numpy()
car_names = df['car name'].to_numpy()
more_than_6_cylinders= cylinder_array<6
car_name = car_names[more_than_6_cylinders]
car_names_list = car_name.tolist()
```

```python
#3. statistical analyse
weight_array = df['weight'].to_numpy()
percentiles = np.percentile(weight_array, [25, 50, 75])
percentiles
```

```python
#4 Array manipulation
acceleration_array = df['acceleration'].to_numpy()

acceleration_min = acceleration_array.min()
acceleration_max = acceleration_array.max()
acceleration_normalized = (acceleration_array - acceleration_min) / (acceleration_max - acceleration_min)

acceleration_normalized[:10]
```

```python
#5.
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Fill missing values with the mean of the column
horsepower_mean = df['horsepower'].mean()
horsepower_filled = df['horsepower'].fillna(horsepower_mean)

# Convert to NumPy array and increase all values by 10%
horsepower_array = horsepower_filled.to_numpy()
horsepower_increased = horsepower_array * 1.10

horsepower_increased[:10]  
```

```python

# 6. Indexing
origin_array = df['origin'].to_numpy()
displacement_array = df['displacement'].to_numpy()

# Filter displacement values for cars with origin == 2 (Europe)
european_displacement = displacement_array[origin_array == 2]

# Compute the average displacement for European cars
average_displacement_europe = np.mean(european_displacement)
average_displacement_europe
```

```python
# 7.Matrix operation
df['horsepower'] = pd.to_numeric(df['horsepower'])
mpg_array = df['mpg'].to_numpy()
horsepower_array = df['horsepower'].to_numpy()
weight_array = df['weight'].to_numpy()


data_matrix = np.column_stack((mpg_array, horsepower_array, weight_array))

vector = np.array([1, 0.5, -0.2])

dot_product_result = np.dot(data_matrix, vector)

dot_product_result[:10]
```

```python
#8.sorting
model_year_array = df['model year'].to_numpy()

sorted = np.argsort(-model_year_array)

sorted_car_names = df['car name'].to_numpy()[sorted]

sorted_car_names[:5]
```

```python
#9.Corelation
mpg_array = df['mpg'].to_numpy()
weight_array = df['weight'].to_numpy()

correlation_matrix = np.corrcoef(mpg_array, weight_array)
pearson_correlation = correlation_matrix[0, 1]
pearson_correlation
```

```python
#10.conditional Aggregates.
cylinders = df['cylinders'].to_numpy()
mpg = df['mpg'].to_numpy()


unique_cylinders = np.unique(cylinders)

mean_mpg_by_cylinder = np.array([
    [cyl, np.mean(mpg[cylinders == cyl])]
    for cyl in unique_cylinders
])
mean_mpg_by_cylinder
```

```python
#2.Filtering and Indexing

car_df = df[(df['model year'] == 75) & (df['weight'] < 3000)]


result_df = car_df[['car name', 'weight', 'mpg']]
result_df
```

```python
#3.Handling Missing Data
missing_values = df.isnull().sum()

missing_values
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'])
median_hp = df['horsepower'].median()
df['horsepower'].fillna(median_hp)

no_missing_values = df.isnull().sum()
no_missing_values
```

```python
#4.Data Transformation
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
df.head(5)
```

```python
#5.
# Group by 
mean_mpg_by_origin = df.groupby('origin')['mpg'].mean()
mean_mpg_by_origin
```

```python
#6.Sorting

top_10_mpg_cars = df.sort_values(by='mpg', ascending=False).head(10)

top_10_mpg_cars[['car name', 'mpg']].reset_index()
```

```python
#7.Apply function

def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']

df['performance_score'] = df.apply(performance_score, axis=1)

df[['car name', 'mpg', 'acceleration', 'weight', 'performance_score']].head()
```

```python
#8. Visualisation preparation

summary_dataframe = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()

summary_dataframe.head()
```

```python
#9.Exporting

mpg_subset = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]

mpg_subset.to_csv("high_mpg_cars.csv")
```

```python
#10.
##IQR 
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1

IQR
```

```python
# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


mpg_outliers_df = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]
mpg_outliers_df
```
